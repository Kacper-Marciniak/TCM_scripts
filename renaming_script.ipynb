## SET DIRECTORY, LOAD FILES

```python
COPY = False

# import os
import os
# assign directory
working_dir = os.path.dirname(os.path.realpath("__file__"))
input_files_dir =  os.path.join(working_dir,'skany 2D')
output_files_dir =  os.path.join(working_dir,'files_out')
output_files_dir = r"J:/skany_przerobione"

# create var to hold all filenames
list_filenames = list()
# create var to hold all folders
list_folders = list()

# get folders
for filename in os.listdir(input_files_dir):
    file = os.path.join(input_files_dir, filename)
    # checking if it is a file
    if not os.path.isfile(file):
        # is a folder
        list_folders.append(str(filename))
        list_filenames.append(list())

if len(list_folders) == 0: list_folders.append("") # no folders inside the working directory

# load files
for i,folder in enumerate(list_folders):
    folder = os.path.join(input_files_dir, folder)
    for filename in os.listdir(folder):
        file = os.path.join(folder, filename)
        # checking if it is a file
        if os.path.isfile(file) and file.endswith(".png") and not "3D" in filename and len(filename) == 11:
            # is a file
            list_filenames[i].append(str(filename.replace(".png","")))

# print info about files and folders inside working directory
print(f"List of folders inside {working_dir}:")
for i,folder in enumerate(list_folders):
    print(f"{i} - > {folder}: N. of files: {len(list_filenames[i])}")   
    
```

## CREATE MATRIX FOR EACH FILE LIST

```python
import numpy as np
list_filenames_new = list()
for list_files in list_filenames:
    list_files.sort()
    list_files_new = list()

    list_rows = list()
    list_sections = list()

    for i,element in enumerate(list_files):
        if not element.split("_")[1] in list_rows:
            list_rows.append(element.split("_")[1])
        if not element.split("_")[0] in list_sections:
            list_sections.append(element.split("_")[0])

    list_rows.sort()
    list_sections.sort()

    matrix_broach = np.empty((len(list_sections), len(list_rows)))

    for i,element in enumerate(list_files):
        r = list_rows.index(element.split("_")[1])
        s = list_sections.index(element.split("_")[0])
        list_files_new.append(f"{s+1:d}_{len(list_rows) - r:d}")  

    list_filenames_new.append(list_files_new)
```

## RENAME FILES

```python
import shutil
from datetime import datetime

file_log = open(os.path.join(working_dir,"log.txt"),"w")
file_log.write(str(datetime.now().strftime("%d/%m/%Y %H:%M:%S"))+"\n")

for i,folder in enumerate(list_folders):
    if not os.path.exists(os.path.join(output_files_dir, folder)):
        os.makedirs(os.path.join(output_files_dir, folder))
    file_log.write(f"Inside: {os.path.join(input_files_dir, folder)}\n")
    for j,filename in enumerate(list_filenames[i]):
        print(f"{i+1}/{len(list_folders)}   {j+1}/{len(list_filenames[i])}")
        path_old = os.path.join(os.path.join(input_files_dir, folder),f"{list_filenames[i][j]}.png")
        path_new = os.path.join(os.path.join(output_files_dir, folder),f"{list_filenames_new[i][j]}.png")
        if COPY: shutil.copy(path_old, path_new)
        else: 
            path_new = os.path.join(os.path.join(input_files_dir, folder),f"{list_filenames_new[i][j]}.png")
            os.rename(path_old, path_new)
        file_log.write(f"{list_filenames[i][j]}.png -> {list_filenames_new[i][j]}.png\n")

file_log.close()
```
